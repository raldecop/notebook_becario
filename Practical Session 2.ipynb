```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
sns.set(rc={"figure.figsize": (10, 5)})
```

```python
df = pd.read_csv('email50.csv')
```

## 1. Robust Statistics

## 2 Transforming Data

```python
plt.scatter(df['num_char'], df['line_breaks'])
plt.xlabel('Number of characters (in thousands)')
plt.ylabel('Number of lines')
```

### Logarithmic transformation:

```python
plt.scatter(np.log(df['num_char']), np.log(df['line_breaks']))
plt.xlabel('Log(Number of characters (in thousands))')
plt.ylabel('Log(Number of lines)')
```

### Square root

```python
plt.scatter(np.sqrt(df['num_char']), np.sqrt(df['line_breaks']))
plt.xlabel('Sqrt(Number of characters (in thousands))')
plt.ylabel('Sqrt(Number of lines)')
```

### Square root inverse

```python
plt.scatter(1/np.sqrt(df['num_char']), 1/np.sqrt(df['line_breaks']))
plt.xlabel('1/Sqrt(Number of characters (in thousands))')
plt.ylabel('1/Sqrt(Number of lines)')
```
